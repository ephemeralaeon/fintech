````python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from collections import OrderedDict 
from statistics import mean
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []
row_count = 0

# @TODO: Read in the menu data into the menu list
with menu_filepath.open() as csvfile:
    spreadsheet = csv.reader(csvfile, delimiter=',')
    next(spreadsheet)
    for menu_list in spreadsheet:
        menu.append(menu_list)
#     print menu

# @TODO: Read in the sales data into the sales list
with sales_filepath.open() as csvfile:
    spreadsheet = csv.reader(csvfile, delimiter=',')
    next(spreadsheet)
    for sales_list in spreadsheet:
        sales.append(sales_list)

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = OrderedDict()

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object
for sales_row in sales:

    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
    Line_Item_ID = int(sales_row[0])
    Date = sales_row[1]
    Credit_Card_Number = int(sales_row[2])
    Quantity = int(sales_row[3])
    Menu_Item = sales_row[4]
    

    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
    if Menu_Item in report.keys():
        pass
    else:
        report[Menu_Item] = OrderedDict()
        report[Menu_Item]["01-count"] = 0
        report[Menu_Item]["02-revenue"] = 0
        report[Menu_Item]["03-cogs"] = 0
        report[Menu_Item]["04-profit"] = 0

    # @TODO: For every row in our sales data, loop over the menu records to determine a match
    for menu_row in menu:
        # Item,Category,Description,Price,Cost
        # @TODO: Initialize menu data variables
        Item = menu_row[0]
        Category = menu_row[1]
        Description = menu_row[2]
        Price = float(menu_row[3])
        Cost = float(menu_row[4])
        
        # @TODO: Calculate profit of each item in the menu data
        Profit = Price - Cost
        # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item
        if Item == Menu_Item:
            # @TODO: Print out matching menu data
#             print menu_row
            # @TODO: Cumulatively add up the metrics for each item key
            report[Menu_Item]["01-count"] += Quantity
            report[Menu_Item]["02-revenue"] += Price * Quantity
            report[Menu_Item]["03-cogs"] += Cost * Quantity
            report[Menu_Item]["04-profit"] += Profit * Quantity
        # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
        else:
            pass
    # @TODO: Increment the row counter by 1
    row_count += 1


# @TODO: Print total number of records in sales data
print row_count
# report = dict(report)
# @TODO: Write out report to a text file (won't appear on the command line output)
f= open("report.txt","w+")
for item in list(report.keys()):
    item_name = item
    info_list = list(report[item])
#     item_info = str(report[item][0] + report[item][1] + report[item][2] + report[item][3])
    f.write(str(item_name) + ':')
    for listitem in info_list:
        f.write(" " + str(listitem) + ": " + str(report[item][listitem])+"; ")
    f.write("\n")   
f.close()
````
